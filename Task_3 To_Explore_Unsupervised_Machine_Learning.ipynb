```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import KMeans
%matplotlib inline
sns.set_style("darkgrid")
```

### Load the data set

```python
df = datasets.load_iris()
df = pd.DataFrame(df.data, columns = df.feature_names)
```

### View the dataset

```python
df.head(5)
```

### Describe the dataset

```python
df.describe()
```

```python
plt.scatter(x = 'sepal length (cm)', y = 'sepal width (cm)', data = df)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
plt.scatter(x = 'petal length (cm)', y = 'petal width (cm)', data = df)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

# Using Elbow method to find n_clusters

```python
wcss_score=[]
k_range = 15
for i in range(1,k_range):
    kmeans = KMeans(i)
    kmeans.fit(df)
    w = kmeans.inertia_
    wcss_score.append(w)
print(wcss_score)
```

```python
clusters=range(1,k_range)
plt.plot(clusters,wcss_score,  marker='.', color = "r", markersize=10)
plt.xlabel('No of clusters', fontsize=15)
plt.ylabel('WCSS', fontsize=15)
plt.title('The elbow method', fontsize=25)
plt.show()
```

# K= 2 clusters

```python
kmeans_2=KMeans(2)
kmeans_2.fit(df)
```

```python
iris_pred_2 = df.copy()
iris_pred_2['predicted'] = kmeans_2.fit_predict(df)
```

### Visualizing clusters according to predicted values

```python
plt.scatter(x='sepal length (cm)', y='sepal width (cm)', c='predicted', cmap = 'RdYlGn', data=iris_pred_2)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('K=2 Sepal length vs sepal width')
plt.show()
```

```python
plt.scatter(x='petal length (cm)', y='petal width (cm)', c='predicted', cmap = 'jet', data=iris_pred_2)
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.title('K=2 petal length vs petal width')
plt.show()
```
